```python
# Veri setindeki null deerleri tespit eder, say ve yzde olarak sralar

def missing_values(df):
    missing_number = df.isnull().sum().sort_values(ascending = False)
    missing_percent = (df.isnull().sum() / df.isnull().count()).sort_values(ascending = False)
    missing_values = pd.concat([missing_number, missing_percent], axis = 1, keys = ['Missing_Number', 'Missing_Percent'])
    return missing_values[missing_values['Missing_Number'] > 0]
```

**Dorusal modeller,** aykr deerlere  hassastr nk bu modeller veri noktalarn en iyi uyan bir izgi (veya daha yksek boyutlu uzaylarda bir dzlem veya hiper dzlem) bulmaya alr. Aykr deerler, modelin bu "en iyi uyum" hedefini byk lde etkileyebilir. te bunun baz nedenleri:

1. **MSE Temelli Kayp Fonksiyonlar**: Dorusal regresyon gibi birok dorusal model, ortalama karesel hata (MSE) temelli bir kayp fonksiyonu kullanr. MSE, byk hatalarn karesini alarak bu hatalar daha da bytr, bu nedenle tek bir aykr deer bile toplam hatay byk lde artrabilir.

2. **Dk Basknlk**: Dorusal modellerde, her bir veri noktas modelin eitilmesine katkda bulunur. Ancak, aykr bir deer, dier tm deerlere kyasla ok daha byk bir etkiye sahip olabilir. Bu, modelin bu aykr deere ar ekilde uymasna neden olabilir.

3. **Dorusal Varsaym**: Dorusal modeller, veriler arasndaki ilikinin dorusal olduunu varsayar. Bu, modelin aykr deerlere zellikle duyarl olmasna neden olabilir, nk bu aykr deerler genellikle varsaylan dorusal ilikiyi bozar.

4. **Kaldra (Leverage) Etkisi**: Aykr deerler, dorusal bir modelin eilimini (veya eimini) byk lde deitirebilir, bu da dier tm veri noktalar iin tahminlerin yanltc olmasna neden olabilir.

Birok uygulamada, dorusal modellerin aykr deerlere olan bu duyarll, modelin genelleme kabiliyetini ve gvenilirliini azaltabilir. Bu nedenle, dorusal modelleme yapmadan nce aykr deerleri tespit etmek ve uygun bir ekilde ele almak genellikle nerilir.

Birok istatistiksel ve makine renimi modeli aykr deerlere duyarldr. Ancak bu duyarllk, modelin zelliklerine ve kullanlan optimizasyon tekniklerine bal olarak deiebilir.

**Aykr Deerlere Duyarl Modeller:**

1. **Dorusal Regresyon**: Daha nce bahsettiimiz gibi, dorusal regresyon aykr deerlere duyarldr.
2. **Lojistik Regresyon**: Lojistik regresyon, dorusal regresyonun bir trdr ve benzer nedenlerle aykr deerlere duyarl olabilir.
3. **K-En Yakn Komu (KNN)**: KNN algoritmas, k deeri kk olduunda aykr deerlerden etkilenebilir.
4. **Tek Deikenli Zaman Serileri Modelleri**: ARIMA gibi modeller aykr deerlere duyarl olabilir.

**Aykr Deerlere Duyarl Olmayan Modeller:**

1. **Decision Tree & Random Forest**: Bu modeller, veri noktalarn blerek alt iin genellikle aykr deerlere kar dayankldr.

## Multicollinearity control

- Multicollinearity sorunu linear ve logisitc reg. gibi gradient descent tabanl modellerde bir problem olarak karmza kar. 

- Multicollinearity bamsz deikenler arasnda yksek bir korelasyon olduunda ortaya kar. Bu durum, bir bamsz deikenin dier bamsz deikenlerle gl bir ekilde ilikili olduunda, dier bamsz deikenlerin modeldeki nemini azaltabilir.

- Multicollinearity sorunu, modelin doru bir ekilde yorumlanmasn zorlatrabilir. Model katsaylarnn doru bir ekilde tahmin edilmesi zordur ve baz deikenlerin etkisinin yanl yorumlanmasna neden olabilir. Bu nedenle, doru sonular elde etmek iin, multikolinerite sorunu ortadan kaldrlmaldr.

- Regulirazition algoritmalar overfiting problemini giderdii gibi Multicollinearity sorununu da giderir.

oklu dorusallk (multicollinearity), bamsz deikenler arasnda yksek derecede dorusal bir iliki olmas durumunu tanmlar. Multicollinearity, baz istatistiksel modeller iin sorun oluturabilirken, bazlar iin bu kadar kritik olmayabilir.

**Multicollinearity Duyarl Modeller:**

1. **Dorusal Regresyon**: Dorusal regresyon modeli, bamsz deikenler arasnda yksek korelasyon (multicollinearity) olduunda ciddi ekilde etkilenebilir. Bu, regresyon katsaylarnn tahminlerinin istikrarsz olmasna ve sonularn yanltc olmasna neden olabilir, nk model, bamsz deikenlerin etkilerini birbirinden ayrtrmakta zorlanr.

2. **Lojistik Regresyon**: Lojistik regresyon da benzer ekilde multicollinearity'den etkilenebilir nk bu da bir tr dorusal modeldir ve bamsz deikenler arasndaki yksek korelasyon, parametre tahminlerinin gvenilirliini azaltabilir.

3. **Ridge ve Lasso Regresyonu**: Ridge ve Lasso regresyonlar, multicollinearity sorununa kar dayankllk salayan dzenliletirme (regularization) teknikleridir. Ancak, bu tekniklerin etkinlii, lambda (dzenliletirme parametresi) deerinin doru seimine baldr. Lambda'nn seimi modelin performans zerinde nemli bir etkiye sahiptir ve bu seim genellikle apraz dorulama yoluyla yaplr.

**Multicollinearity Duyarl Olmayan Modeller:**

1. **Karar Aalar**: Karar aalar ve karar aalarna dayal modeller (rnein, Random Forest), bamsz deikenler arasndaki ilikilerden etkilenmezler nk bu modeller blnme kriterlerine dayanarak alr ve her bir bamsz deikeni ayr ayr deerlendirir.

2. **Destek Vektr Makineleri (SVM)**: SVM, zellikle doru kernel fonksiyonu seildiinde, bamsz deikenler arasndaki yksek korelasyondan etkilenmez. Model, veri noktalarn optimum bir hiperdzlemle ayrma yeteneine sahiptir ve bu srete deikenler arasndaki korelasyon nemli bir sorun tekil etmez.

3. **Naive Bayes**: Naive Bayes snflandrcs, deikenler arasndaki ilikilere dayanmaz ve bamszlk varsaym altnda alr, bu nedenle multicollinearity'den etkilenmez.

**Multicollinearity Neden Sorun Tekil Eder?**

- **Katsaylarn Yorumlanabilirlii**: Multicollinearity varlnda, bamsz deikenlerin birbirleriyle yksek derecede korelasyon gstermesi, bu deikenlerin model zerindeki etkisinin doru bir ekilde yorumlanmasn zorlatrr.
  
- **Tahminlerin stikrarszl**: Kk veri deiiklikleri, model katsaylarnda byk dalgalanmalara neden olabilir, bu da modelin tahmin gvenilirliini azaltr.
  
- **Test statistiklerinin Gvenilmezlii**: Multicollinearity, istatistiksel testlerin sonularn yanltc hale getirebilir, rnein, p-deerlerinin ve R-kare deerlerinin yanltc olmas gibi.

- Multicollinearity ile baa kmak iin deiken seimi, dzenliletirme teknikleri veya ana bileenler regresyonu gibi yntemler kullanlabilir.

- Bu yntemler, modelin tahmin gvenilirliini artrarak ve katsaylarn yorumlanabilirliini iyiletirerek multicollinearity'nin olumsuz etkilerini azaltmaya yardmc olabilir.

# Implement Linear Regression

## Model

```python
def train_val(model, X_train, y_train, X_test, y_test):
    
    y_pred = model.predict(X_test)
    y_train_pred = model.predict(X_train)
    
    scores = {
    "train": {
    "R2" : r2_score(y_train, y_train_pred),
    "mae" : mean_absolute_error(y_train, y_train_pred),
    "mse" : mean_squared_error(y_train, y_train_pred),                          
    "rmse" : np.sqrt(mean_squared_error(y_train, y_train_pred))},
    
    "test": {
    "R2" : r2_score(y_test, y_pred),
    "mae" : mean_absolute_error(y_test, y_pred),
    "mse" : mean_squared_error(y_test, y_pred),
    "rmse" : np.sqrt(mean_squared_error(y_test, y_pred))}
               }
    
    return pd.DataFrame(scores)
```

```python
X.columns
```

```python
column_trans = make_column_transformer((OneHotEncoder(handle_unknown="ignore",drop="first",
                                                     sparse_output=False),['DESIGNATION', 'UNIT']),
    remainder=StandardScaler())


operations = [('column_trans',column_trans), ('lm_model', LinearRegression())]
                                              
lm_pipe = Pipeline(steps= operations)

lm_pipe.fit(X_train, y_train)
```

```python
train_val(lm_pipe, X_train, y_train, X_test, y_test)
```

```python
y_pred = lm_pipe.predict(X_test)
lm_R2 = r2_score(y_test, y_pred)
lm_mae = mean_absolute_error(y_test, y_pred)
lm_rmse = np.sqrt(mean_squared_error(y_test, y_pred))
```

## Adjusted R2 Score

```python
def adj_r2(y_test, y_pred, df):
    r2 = r2_score(y_test, y_pred)       # Modelin R2 deerini hesaplar
    n = df.shape[0]                     # Veri setindeki gzlem (rnek) saysn alr
    p = df.shape[1]-1                   # Bamsz deikenlerin (feature larn) saysn alr 
    adj_r2 = 1 - (1-r2)*(n-1)/(n-p-1)   # Adjusted R2 formln kullanarak deeri hesaplar
    return adj_r2
```

```python
y_pred = lm_pipe.predict(X_test)
```

```python
adj_r2(y_test, y_pred, df)
```

```python
# tm modellerden elde ettiimiz metricleri scores deikenine atyoruz. Daha sonra model isimlerinin
# indexte, metriclerin featurlarda grnmesi maksadyla df'in transposunu alyoruz.

scores = {
    "linear_m": {
        "r2_score": lm_R2,
        "mae": lm_mae,
        "rmse": lm_rmse
    },
    
     "linear3_m": {
        "r2_score": lm3_R2,
        "mae": lm3_mae,
        "rmse": lm3_rmse
     },
    
    "ridge_m": {
        "r2_score": rd_R2,
        "mae": rd_mae,
        "rmse": rd_rmse
    },
    "lasso_m": {
        "r2_score": ls_R2,
        "mae": ls_mae,
        "rmse": ls_rmse
    },
    "elastic_m": {
        "r2_score": el_R2,
        "mae": el_mae,
        "rmse": el_rmse
    },
    "xgb_m": {
        "r2_score": xgb_R2,
        "mae": xgb_mae,
        "rmse": xgb_rmse
    },
     "rf_m": {
        "r2_score": rf_R2,
        "mae": rf_mae,
        "rmse": rf_rmse
    }
}
scores = pd.DataFrame(scores).T
scores
```

```python
#metrics = scores.columns

for i, j in enumerate(scores):
    plt.figure(i)
    if j == "r2_score":
        ascending = False # eer metricimiz r2_score ise barplot bykten ke sralanacak
    else:
        ascending = True # eer metricimiz mae veya rmse ise barplot kkten bye sralanacak
    compare = scores.sort_values(by=j, ascending=ascending)
    ax = sns.barplot(x = compare[j] , y= compare.index)
    for p in ax.patches:
            width = p.get_width()                        # get bar length
            ax.text(width,                               # set the text at 1 unit right of the bar
                    p.get_y() + p.get_height() / 2,      # get Y coordinate + X coordinate / 2
                    '{:.4f}'.format(width),             # set variable to display, 2 decimals
                    ha = 'left',                         # horizontal alignment
                    va = 'center') 
```
